```python
def get_browser():
    chrome_options = wd.ChromeOptions()
    
    browser = wd.Chrome(options=chrome_options)
    return browser
```

```python
def translate(comment):
    base_url = 'https://translate.google.com/#'
    from_lang = 'auto'
    to_lang = 'en'
    final_url = base_url + from_lang + '/' + to_lang
    browser.get(final_url)
    time.sleep(1)
    browser.find_element_by_id('source').send_keys(str(comment).lower())
    time.sleep(2)
    text_output = browser.find_elements_by_xpath("/html/body/div[2]/div[2]/div[1]/div[2]/div[1]/div[1]/div[2]/div[3]/div[1]/div[2]/div/span[1]")
    out = ''
    for i in text_output:
        out = i.text
    print(out)
    time.sleep(2)
    return out
```
